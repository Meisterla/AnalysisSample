```python
df['anime_year_all'] = df.anime_year.astype(str) + df.anime_year_second.astype(str)
```

```python
df = df.drop(['anime_year','anime_year_second'],axis=1)
```

```python
df.anime_year_all = df.anime_year_all.str.replace('nan','')
```

```python
company_info = df[['anime_company','anime_name','anime_set_number','anime_year_all','anime_cast']]
```

```python
company_info
```

```python
company_info=company_info.drop('anime_year_all', axis=1).join(company_info['anime_year_all'].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename('year_temp')).reset_index(drop = True)
```

```python
company_info=company_info.drop('year_temp', axis=1).join(company_info['year_temp'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).rename('year_temp2')).reset_index(drop = True)
```

```python
company_info=company_info.drop('year_temp2', axis=1).join(company_info['year_temp2'].str.split('，', expand=True).stack().reset_index(level=1, drop=True).rename('year')).reset_index(drop = True)
```

```python
company_info=company_info.drop('anime_cast', axis=1).join(company_info['anime_cast'].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename('cast')).reset_index(drop = True)
```
